The datasets above don't give us enough information because the column names and values are coded from survey data. We need to drop irrelevant columns and rename the columns to something we can understand, and recode the observation value data back to its original description. The dataset from Kaggle came with a codebook, so we were able to reference the columns and create dictionaries from the survey description.

```python
#selecting only relevant columns and demographic/socioeconomic factors info
dem = dem[['SEQN', 'DMDBORN4','DMDEDUC2','DMDHHSIZ','RIDAGEYR','INDFMPIR','INDHHIN2','RIAGENDR','RIDRETH3']]
dem
```

```python
#selecting only relevant cumns from insurance data set
insurance = insurance[['SEQN', 'HIQ011','HUQ051']]
insurance
```

```python
#merge data sets into aggregated data set according to the individual survey response
df = pd.merge(dem, insurance, on='SEQN', how='left')
df
```

```python
#rename columns
df = df.rename(columns={
    'SEQN': 'respondent_id',
    'DMDBORN4': 'usa_birth',
    'DMDEDUC2': 'education',
    'DMDHHSIZ': 'household_size',
    'RIDAGEYR': 'age',
    'INDFMPIR': 'poverty_ratio',
    'INDHHIN2': 'income',
    'RIAGENDR': 'gender',
    'RIDRETH3': 'race',
    'HIQ011': 'insurance',
    'HUQ051': 'num_healthcare'
})

df
```

```python
#recode values from survey data to actual info
#values obtained from data set and codebook files with code/value and associated descriptions

#race
race_dict = {
    1: 'Hispanic',
    2: 'Hispanic',
    3: 'White',
    4: 'Black',
    6: 'Asian' ,
    7: 'Other'}

df['race'].replace(race_dict, inplace=True)      
```

```python
#usa_birth
#if born in us, keep value as 1, if other, return 0 for false
usa_dict = {
    2: '0'}

df['usa_birth'].replace(usa_dict, inplace=True)
df
```

```python
#highest level of education
education_dict = {
    1: '<9th grade',
    2: '9th-11th grade',
    3: 'high school',
    4: 'some college',
    5: 'college or above'
}

df['education'].replace(education_dict, inplace=True)
df = df[df['education'].isin(set(education_dict.values()))]


df
```

```python
#annual income
                                   
income_dict = {
    1: '0-4999',
    2: '5,000-9,999',
    3: '10,000-14,999',
    4: '15,000-19,999',
    5: '20,000-24,999',
    6: '25,000-34,999',
    7: '35,000-44,999',
    8: '45,000-54,999',
    9: '55,000-64,999',
    10:'65,000-74,999',
    14:'75,000-99,999',
    15:'> 100,000'
    
}

df['income'].replace(income_dict, inplace=True)
df = df[df['income'].isin(set(income_dict.values()))]
                                   
df
```

```python
#insurance
insurance_dict = {
    2 : '0'}

df['insurance'].replace(insurance_dict, inplace=True)
```

```python
#gender
gender_dict = {
    1 : 'male',
    2 : 'female'}

df['gender'].replace(gender_dict, inplace=True)
```

```python
#organized dataset
#subset data to only include age groups 26 to 34 years old and 35 to 49 years old
df = df[(df['age'] >= 25) & (df['age'] <= 49)]
df
#remove na and null values

```

## Hypothesis Testing

After exploring some of the initial relationship and analysis of healthcare insurance and determinants like education level and poverty, ratio, we wanted to conduct some hypothesis testing to see if our thoughts and hypotheses hold up, We decided to explore this with chi square (x^2) hypothesis testing, because are mainly looking at categorical variables, and other methods with z scores or linear regression wouldn't have worked in our case.

First, lets take a look at a correlation matrix table to see the associated relationship between healthcare insurance status, age, poverty ratio, and the number of times someone visits the hospital and receives healthcare assistance.

```python
numeric_col = ['age','poverty_ratio','num_healthcare','insurance']

corr_matrix = df.loc[:,numeric_col].corr()
print(corr_matrix)
 
#Using heatmap to visualize the correlation matrix
sns.heatmap(corr_matrix, annot=True)
```

We can see that poverty ratio has a moderately positive relationship with health insurance coverage. On the other hand, age does not have much of a strong relationship with health insurance status.

Now, we'll conduct 2 variable chi square hypothesis testing with insurance and an interaction variable: age, gender, poverty ratio, and education.

```python
contingency_table = pd.crosstab(df['insurance'], df['race'])

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square value: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
```

With a df of 4 and chi square value of 102, which is greater than the chi square critical value of 9.488. Because of this, we reject the null hypothesis that race is not an indicator of health insurance and conclude that race is a factor in whether a person has access to health insurance or not.

```python
contingency_table1 = pd.crosstab(df['insurance'], df['gender'])

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table1)
print(f"Chi-square value: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
```

We fail to reject the null hypothesis because the chi square value is smaller than the critical value, even if we have a p value of .001, which is smaller than alpha value at 0.05 level of significance. This indicates that knowing the gender does not inform us of their healthcare insurance status.

```python
contingency_table2 = pd.crosstab(df['insurance'], df['poverty_ratio'])

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table2)
print(f"Chi-square value: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
```

With a p value less than 0.05 and a chi square value greater than the critical value of 411.610, we reject the null hypothesis that a person's family poverty ratio is not indicative of healthcare insurance status and conclude that it does in fact inform us if a person has access to healthcare insurance.
